```python
import pandas as pd
raw_data = pd.read_csv("test_result.csv")
```

```python
raw_data
```

```python
raw_data.index +=1
```

```python
train_data = raw_data.iloc[:,1:9]
```

```python
from sklearn.cluster import KMeans 
num_clusters = 2
model = KMeans(init='k-means++', n_clusters=num_clusters, n_init=50)
model.fit(train_data)
```

```python
y = model.predict(train_data)
train_data["cluster"]=y
Group_A = list(train_data[train_data["cluster"]==1].index)
print(f"Group_A = {Group_A}")
```

```python
Group_B = list(train_data[train_data["cluster"]==0].index)
print(f"Group_B = {Group_B}")
```

# As we can see, the grouping is almost perfect, as most of the science stream students are classified into the same group.
# Also, most art students are classified into another group.
